### Задание 1

```python
def is_leap_year(year: int) -> str:
    is_leap = year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)
    return is_leap and 'Високосный год' or 'Обычный год'
```

```python
is_leap_year(2020)
```

```python
is_leap_year(2019)
```

### Задание 2

```python
def is_lucky_ticket(ticket_number) -> str:
    digits = [int(digit) for digit in str(ticket_number)]
    is_lucky = sum(digits[:3]) == sum(digits[-3:])
    return is_lucky and 'Счастливый билет' or 'Несчастливый билет'
```

```python
is_lucky_ticket(123456)
```

```python
is_lucky_ticket(123321)
```
