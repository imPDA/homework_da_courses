### Задание 1

```python
from datetime import datetime, timedelta
```

```python
def parse_moscow_date(date_str):
    return datetime.strptime(date_str, '%A, %B %d, %Y')

def parse_guardian_date(date_str):
    return datetime.strptime(date_str, '%A, %d.%m.%y')

def parse_daily_news_date(date_str):
    return datetime.strptime(date_str, '%A, %d %B %Y')
```

```python
parse_moscow_date('Wednesday, October 2, 2002')
```

```python
parse_guardian_date('Friday, 11.10.13')
```

```python
parse_daily_news_date('Thursday, 18 August 1977')
```

### Задание 2

```python
def date_range(start, end):
    try:
        start_dt = datetime.strptime(start, '%Y-%m-%d')
        end_dt = datetime.strptime(end, '%Y-%m-%d')
    except ValueError:
        return []
    
    if start_dt > end_dt:
        return []
    
    days = (end_dt - start_dt).days + 1

    return [(start_dt + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(days)]
```

```python
date_range('2022-01-01', '2022-01-03')
```

```python
date_range('2022-01-03', '2022-01-01')
```

```python
date_range('2022-02-30', '2022-02-31')
```
