```python
years = list(map(str, range(1950, 2011)))


def production_year(title: str) -> int:
    for year in years:
        if year in title:
            return int(year)
    
    return 1900
```

```python
import pandas as pd
from pathlib import Path


BASE_DIR = Path('/home/jovyan/work/homework') 
MOVIES_PATH = BASE_DIR / 'data/movies_stats.csv'

movies_df = pd.read_csv(MOVIES_PATH)
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df['year'] = movies_df['title'].apply(production_year)
movies_df.head()
```

```python
rating_vs_year = movies_df.groupby('year')['rating'].mean()
rating_vs_year.sort_values(ascending=False, inplace=True)

rating_vs_year
```
