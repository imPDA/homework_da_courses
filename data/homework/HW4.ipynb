```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

### Задание 1 и 2

```python
# задание 1
def search_name(number):
    for document in documents:
        if document['number'] == number:
            return f'Владелец документа: {document['name']}'
    else:
        return 'Владелец документа: владелец не найден'


# задание 2
def search_shelf(number):
    for directoryIndex, directoryDocuments in directories.items():
        if number in directoryDocuments:
            return f'Документ хранится на полке: {directoryIndex}'
    else:
        return 'Документ не найден'
```

```python
while True:
    command = input('Введите команду:')
    if command == 'q':
        break

    argument = input('Введите номер документа:')

    if command == 'p':
        print(search_name(argument))
    elif command == 's':
        print(search_shelf(argument))
    else:
        print('Неизвестная команда')
```
