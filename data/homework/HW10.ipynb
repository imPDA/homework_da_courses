### Задание 1

```python
import pandas as pd
from pathlib import Path


BASE_DIR = Path('/home/jovyan/work/homework')
MOVIES_PATH = BASE_DIR / 'data/movies/movies.csv'
RATINGS_PATH = BASE_DIR / 'data/movies/ratings.csv'
```

```python
ratings = pd.read_csv(RATINGS_PATH)
ratings.head()
```

```python
ratings.dtypes
```

```python
RATING_FIVE = ratings.rating == 5

sorted_by_rating = ratings[RATING_FIVE].groupby(by='movieId').aggregate('count')['userId'].sort_values(ascending=False)
sorted_by_rating
```

```python
top_rated_movie_id = sorted_by_rating.idxmax()
top_rated_movie_id
```

```python
movies = pd.read_csv(MOVIES_PATH)
movies.head()
```

```python
top_rated_movie = movies[movies['movieId'] == top_rated_movie_id]
top_rated_movie
```

```python
top_rated_movie_name = top_rated_movie['title'].values[0]
print(f'Фильму "{top_rated_movie_name}" было выставлено больше всего оценок 5.0.')
```

### Задание 2

```python
POWER_PATH = BASE_DIR / 'data/transactions/power.csv'

power = pd.read_csv(POWER_PATH)
power.head()
```

```python
power.dtypes
```

```python
power['country'].unique()
```

```python
CATEGORIES = power['category'].isin((4, 12, 21))
COUNTRIES = power['country'].isin(('Latvia', 'Lithuania', 'Estonia'))
YEARS = power['year'].between(2005, 2010, 'both')
POSITIVE_QUANTITY = power['quantity'] > 0

total_consumption = power[CATEGORIES & COUNTRIES & YEARS & POSITIVE_QUANTITY]['quantity'].sum()
print(f'Cуммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год (не учитывая отрицательное потребление): {total_consumption}')
```
