### Задание 1

```python
def get_middle_letters(word: str) -> str:
    word_length = len(word)
    if word_length % 2 == 0:
        return word[word_length//2 - 1:word_length//2 + 1]
    else:
        return word[(word_length-1)//2]
```

```python
word = 'test'
get_middle_letters(word)
```

```python
word = 'testing'
get_middle_letters(word)
```

### Задание 2

```python
def print_pairs(boys: list, girls: list) -> None:
    if len(boys) != len(girls):
        print('Результат: Внимание, кто-то может остаться без пары!')
        return

    boys.sort()
    girls.sort()

    print('Результат:\nИдеальные пары:')
    for boy, girl in zip(boys, girls):
        print(f'{boy} и {girl}')
```

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard'] 
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
print_pairs(boys, girls)
```

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael'] 
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
print_pairs(boys, girls)
```
