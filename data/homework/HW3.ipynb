### Задание 2

```python
from collections import defaultdict
```

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

total = len(queries)
counter = defaultdict(int)
for q in queries:
    counter[len(q.split())] += 1

sorted_counter = sorted(counter.items())

for count in sorted_counter:
    print(f'Поисковых запросов, содержащих {count[0]} слов(а): {count[1] / total:.2%}')
```
