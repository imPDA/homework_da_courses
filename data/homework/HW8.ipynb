### Задание 1

```python
import requests
```

```python
API_URL = 'https://www.cbr-xml-daily.ru/daily_json.js'
```

```python
response = requests.get(API_URL)

response.raise_for_status()

data = response.json()

max_name = ''
max_value = -1
for currency in data['Valute'].values():
    value = currency['Value']

    if value > max_value:
        max_name = currency['Name']
        max_value = value

print(max_name)
```

### Задание 2

```python
from functools import partial


class Rate:
    API_URL = 'https://www.cbr-xml-daily.ru/daily_json.js'

    def __init__(self, diff: bool = False):
        self.diff = diff

    def _request_data(self) -> dict:
        response = requests.get(self.API_URL)
        response.raise_for_status()

        return response.json()['Valute']

    def get_currency_exchange_rate(self, currency_code: str) -> float | None:
        try:
            data = self._request_data()
        except:
            return

        currency_data = data[currency_code]
        if not currency_data:
            return
        
        if self.diff:
            return currency_data['Value'] - currency_data['Previous']
        else:
            return currency_data['Value']

    def __getattr__(self, name: str):
        return partial(self.get_currency_exchange_rate, name.upper())
```

```python
r = Rate(diff=True)
r.usd()
```

```python
r = Rate(diff=False)
r.usd()
```
